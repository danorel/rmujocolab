```python
import mujoco
import mujoco.viewer
import mediapy as media
```

```python
m = mujoco.MjModel.from_xml_path("tutorial/models/box_and_square_joint.xml")
d = mujoco.MjData(m)
```

```python
renderer = mujoco.Renderer(m)

scene_option = mujoco.MjvOption()
scene_option.flags[mujoco.mjtVisFlag.mjVIS_JOINT] = True
```

```python
duration = 5
fps = 60

frames = []
mujoco.mj_resetData(m, d)
while d.time < duration:
    mujoco.mj_step(m, d)
    if len(frames) < d.time * fps:
        renderer.update_scene(d, scene_option=scene_option)
        pixels = renderer.render()
        frames.append(pixels)
media.show_video(frames, fps=fps)
```

```python
print('default gravity', m.opt.gravity)
m.opt.gravity = (0, 0, 10)
print('flipped gravity', m.opt.gravity)
```
